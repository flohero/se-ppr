```python
# import neccessary packages here
import json
import sqlite3
```

### a) JSON to sqlite 3

Write a Script that extracts the data from "newsletter.json" and stores it to an sqlite3 database. How you create the table and transfer the data is up to you.

```python
with open("newsletter.json", "r") as json_file:
    data = json.load(json_file)
con = sqlite3.connect("newsletter.sqlite")
cur = con.cursor()
cur.execute("drop table if exists member")
cur.execute("create table member (id, first_name, last_name, email, gender, ip_address, subscribed_at, premium_member)")
cur.executemany("INSERT INTO member VALUES(?, ?, ?, ?, ?, ?, ?, ?)", [tuple(v.values()) for v in data])
con.commit()
```

### b) SQL statements
Write a script to perform the following statements:

- Get the full names of all the premium users

```python
for row in cur.execute("select first_name, last_name from member where premium_member = true"):
    print(row[0] + " " + row[1])
```

- What is the ratio of premium users?

```python
count_premium = cur.execute("select count(*) from member where premium_member = true").fetchone()
count = cur.execute("select count(*) from member").fetchone()
print(count_premium[0]/count[0])
```

- What is the most common first name?

```python
most_common_first_name = cur.execute("select first_name, count(id) as c from member group by first_name order by c desc").fetchone()
print(most_common_first_name)
```
